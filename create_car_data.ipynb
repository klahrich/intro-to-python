```python
def create_random_data(max_cars=100,
                       model_choices=["chevrolet", "honda", "nissan", "mazda", "ford"],
                       min_timespan=10,
                       max_timespan=1000):
    data_list = []

    for car in range(max_cars):
        x_start = 0
        y_start = 0
        car_id = np.random.randint(1, nb_cars)
        trip_length = np.random.randint(min_timespan, max_timespan)
        data_one_car = pd.DataFrame({'car_id': car_id ,
                                     'timestamp': range(trip_length),
                                     'acceleration_x': np.random.uniform(-1, 1, trip_length),
                                     'acceleration_y': np.random.uniform(-1, 1, trip_length)})

        data_one_car['speed_x'] = data_one_car['acceleration_x'].cumsum().shift(1)
        data_one_car['speed_y'] = data_one_car['acceleration_y'].cumsum().shift(1)

        data_one_car.loc[0, 'speed_x'] = 0
        data_one_car.loc[0, 'speed_y'] = 0

        data_one_car['position_x'] = data_one_car['speed_x'].cumsum().shift(1)
        data_one_car['position_y'] = data_one_car['speed_y'].cumsum().shift(1)

        data_one_car.loc[0, 'position_x'] = 0
        data_one_car.loc[0, 'position_y'] = 0

        data_list.append(data_one_car)

    data_all_cars = pd.concat(data_list, axis=0)
    
    unique_cars = data_all_cars['car_id'].unique()

    data_car_info = pd.DataFrame({'car_id': unique_cars,
                                  'model': np.random.choice(model_choices, len(unique_cars)),
                                  'year': np.random.randint(2000, 2019, len(unique_cars)),
                                  'mileage': np.random.uniform(10000, 100000, len(unique_cars))})
    
    data = pd.merge(data_all_cars, data_car_info, on='car_id')
    
    return data
```
