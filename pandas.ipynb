### Working with text

```python
data.moves.head()
```

```python
data.moves.str.slice(0, 5)
```

```python
import re

data.moves.str.extract("([a-z]{1,2}\d)\s+([a-z]{1,2}\d)", flags=re.IGNORECASE)
```

```python
data[['white_move', 'black_move']] = data.moves.str.extract("([a-z]{1,2}\d)\s+([a-z]{1,2}\d)", flags=re.IGNORECASE)
```

### Group by

```python
data.groupby(['white_move', 'black_move']).size()
```

```python
data.groupby(['white_move', 'black_move']).size().reset_index()
```

```python
data['white_wins'] = (data['winner'] == 'white')
data['black_wins'] = (data['winner'] == 'black')

data[['winner', 'white_wins', 'black_wins']].head()
```

```python
# optional: convert column types
data['white_wins'] = data['white_wins'].astype(int)
data['black_wins'] = data['black_wins'].astype(int)

data[['winner', 'white_wins', 'black_wins']].head()
```

```python
import numpy as np

data.groupby(['white_move', 'black_move']).agg(white_wins=('white_wins', np.sum), black_wins=('black_wins', np.sum)).reset_index()
```

### Concatenating, Merging

```python
players_white = data['white_id']

players_black = data['black_id']

players_both = pd.concat([players_white, players_black])

players_unique = players_both.unique()

players_unique_df = pd.DataFrame({'player_id': players_unique})

players_unique_df
```

```python
when_white_df = pd.merge(players_unique_df, data, left_on='player_id', right_on='white_id', how='inner')
when_black_df = pd.merge(players_unique_df, data, left_on='player_id', right_on='black_id', how='inner')
complete_df = pd.concat([when_white_df, when_black_df], axis=0)
```

```python
len(complete_df)
```

```python
complete_df.sort_values('player_id', inplace=True)
complete_df.head()
```
